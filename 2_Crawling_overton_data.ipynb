## 2. PDF URL Crawling

```python

# Specify the directory path where PDF files will be saved
path = '/content/drive/MyDrive/kdis_python/python project/'
raw_pdf_directory = os.path.join(path, "raw_pdf")

# Function to download a PDF and save it locally
def download_pdf(url, file_name):
    file_path = os.path.join(raw_pdf_directory, file_name)

    # Check if the file already exists
    if os.path.exists(file_path):
        print(f"File '{file_name}' already exists. Skipping download.")
        download_yn = 'Y'
        return download_yn

    try:
        response = requests.get(url)
        response.raise_for_status()  # Check for HTTP request errors
        with open(file_path, "wb") as file:
            file.write(response.content)
        print(f"PDF file successfully downloaded as '{file_name}'")
        download_yn = 'Y'
    except Exception as e:  # Catch all exceptions
        print(f"An error occurred while downloading '{file_name}': {e}")
        download_yn = 'N'
    return download_yn

# Function to process PDF downloads
def download_pdfs(titles, urls):

    download_yn = []
    for i, (title, url) in enumerate(zip(titles, urls), start=1):
        # Sanitize the title to avoid any invalid characters in the file name
        sanitized_title = "".join([c if c.isalnum() or c in "._-" else "_" for c in title])
        pdf_file_name = f"{i}_{sanitized_title}.pdf"

        # Download the PDF
        download_yn.append(download_pdf(url, pdf_file_name))
    return download_yn
```

```python
df = filtered_overton_data.copy()

# Call the function to download PDFs
download_yn = download_pdfs(df['Title'], df['PDF URL'])
```

```python
df['download_yn'] = download_yn
df.to_excel(path+'checking_download_pdf.xlsx')
```

## 4. Extract text from pdf

```python
import os
from PyPDF2 import PdfReader
import pandas as pd
import numpy as np

# Function to extract text from a PDF file
def extract_text_from_pdf(pdf_path):
    try:
        # Open the PDF file
        with open(pdf_path, 'rb') as file:
            reader = PdfReader(file)
            text = ""
            # Extract text from each page
            for page in range(len(reader.pages)):
                text += reader.pages[page].extract_text()
        return text
    except:
        print(f"Error: {pdf_path} - Unable to read PDF file.")
        return None  # Return None in case of an error

# Extract text from all PDF files in a folder
def extract_text_from_folder(folder_path):
    pdf_texts = {}

    # Check all files in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith(".pdf"):  # Process only PDF files
            pdf_path = os.path.join(folder_path, filename)
            pdf_text = extract_text_from_pdf(pdf_path)
            pdf_texts[filename] = pdf_text

    return pdf_texts
```

```python
# Enter the path to the folder containing the PDFs
folder_path = '/content/drive/MyDrive/kdis_python/python_project/raw_pdf'

# Text extraction
pdf_texts = extract_text_from_folder(folder_path)

```

```python
file_path = '/content/drive/MyDrive/kdis_python/python_project/'
file_name = 'checking_download_pdf.xlsx'
```

```python
df_metadata = pd.read_excel(file_path + file_name, index_col=False)
# df_metadata = df_metadata.reset_index()
df_metadata = df_metadata.rename(columns = {'Unnamed: 0':'id'})
df_metadata.id += 1
```

```python
df_data = pd.DataFrame({'file_name' : pdf_texts.keys(), 'text' : pdf_texts.values()})

df_data['id'] = list(map(lambda x:int(x.split('_')[0]), df_data.file_name))
df_data2 = pd.merge(df_metadata, df_data, on = 'id')
df_data2.head(2)
```

```python
df_data2.columns
df_data2.info()
```

```python
import re

# Function to remove illegal characters
def remove_illegal_characters(text):
    ILLEGAL_CHARACTERS_RE = re.compile(r'[\000-\010]|[\013-\014]|[\016-\037]')
    if isinstance(text, str):
        return ILLEGAL_CHARACTERS_RE.sub('', text)
    return text

df_data2['text'] = df_data2['text'].apply(remove_illegal_characters)
```

```python
import pickle
data = df_data2.loc[~df_data2.text.isna()]
file_path = '/content/drive/MyDrive/kdis_python/python_project/'

# Save as pickle file
with open(file_path + 'data.pkl', 'wb') as file:
    pickle.dump(data, file)
```
